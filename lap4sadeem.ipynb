```python
#Q1
print("enter 10 Numbers:")
counter=0
sum=0
while True:
   number=int(input("enter number"))
   if number<0:
      print("can not Read Negative Number:")
      continue
   if number==0:
      break
   if number%2==0:
      sum+=number
   counter+=1
   if counter==10:
      break
print("Sum of Even Numbers = ",sum)
```

```python
#Q2
print("read number to generate multiplication table:")
number=int(input("read number"))
for x in range(1,number+1,1):
    if x==3:
        continue
    product=x*12
    if product>=50:
        break
    print((x*12))
```

```python
#Q3
counter=1
while True:
   number=int(input("Guess number")) 
   if number >=1 and number <=20:
      print ("Guess Correctly: in Attempts :",counter) 
      break
   else:
      counter+=1
      print("Error Guess and Total Attempts Number = ",counter)
      continue
```

```python
#Q4
counter=0
number=int(input("Enter Number :"))
for x in range(1, number + 1, 1):
   if x >50:
      print("Stop Counting ")
      break
   if x %8==0:
      continue
   if x %4==0:
      counter+=1
print("Number Divided by 4=", counter)
```
