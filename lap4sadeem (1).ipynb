```python
#Q1
print("enter 10 Numbers:")
counter=0
sum=0
while True:
   number=int(input("enter number"))
   if number<0:
      print("can not Read Negative Number:")
      continue
   if number==0:
      break
   if number%2==0:
      sum+=number
   counter+=1
   if counter==10:
      break
print("Sum of Even Numbers = ",sum)
```

```python
#Q2
number = int(input())
for i in range(1, 13):
    if i % 3 == 0:
        continue
    product = number * i
    if product > 50:
        break
    print(product)
```

```python
#Q4
counter=0
number=int(input("Enter Number :"))
for x in range(1, number + 1, 1):
   if x >50:
      print("Stop Counting ")
      break
   if x %8==0:
      continue
   if x %4==0:
      counter+=1
print("Number Divided by 4=", counter)
```
